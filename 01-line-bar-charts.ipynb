```python
from matplotlib import pyplot as plt
```

```python
#plt.style.use("fivethirtyeight")
plt.style.use("ggplot")
```

```python
x = [20, 23, 24, 25, 27, 28, 29, 30, 31, 32, 34, 35]
y = [10000, 23000, 20000, 34000, 42000, 43000, 46000, 47000, 50000, 55000, 65000, 72000]
```

```python
py_y = [13000, 17000, 21000, 23000, 26000, 32000, 35000, 40000, 42000, 47000, 50000, 55000]
```

```python
java_y = [11000, 13000, 17000, 20000, 24000, 27000, 32000, 35000, 41000, 47000, 63000, 68000]
```

```python
plt.plot(x, py_y, label = "Python")
plt.plot(x, java_y, label = "JavaScript")
plt.plot(x, y, color = "k", linewidth = "3", label = "All Devs")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
#plt.grid(True)
plt.tight_layout()
plt.savefig("plot.png")
plt.show()
```
