```python
import zipfile
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = r"C:\Users\sukai\Downloads\archive (6).zip"
with zipfile.ZipFile(data) as z:
    with z.open("Iris.csv") as f:
        data1 = pd.read_csv(f)
        id = data1["Id"]
        lengths = data1["SepalLengthCm"]
data1.head()
```

```python
plt.xlabel("Sepal Length CM")
plt.ylabel("Total no. of Sepal Lengths")
plt.title("Iris Dataset")
median_val = lengths.median()
plt.hist(lengths, bins = 5, edgecolor = "black", log = True)
plt.axvline(median_val, color = "yellow", label = "Median Length")
plt.show()
```
